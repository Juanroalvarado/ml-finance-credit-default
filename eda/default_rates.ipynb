```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
full_df=pd.read_csv('train.csv')
full_df.columns
```

```python
len(full_df)
```

```python
full_df= full_df.drop(columns=['Unnamed: 0'])
```

```python
full_df.head()
```

```python
full_df['stmt_date'] = pd.to_datetime(full_df['stmt_date'])
full_df['def_date'] = pd.to_datetime(full_df['def_date'])
full_df.head()
```

```python
# Calculate 1 year after stmt_date
one_year_after = full_df['stmt_date'] + pd.DateOffset(years=1)

# Define default if def_date is between stmt_date and one_year_after
full_df['default'] = ((full_df['def_date'] >= full_df['stmt_date']) & (full_df['def_date'] <= one_year_after)).astype(int)
```

```python
full_df[full_df['default'] == 1]
```

<h1> #Overall default rate for the entire dataset

```python
overall_default_rate = len(full_df[full_df['default'] == 1])/ len(full_df)*100
print('overall default rate:', overall_default_rate,',overall default:',len(full_df[full_df['default'] == 1]),',overall_non_default:',len(full_df[full_df['default'] == 0]))
```

<h1># sectorwise Overall default rate

```python
sector_list=full_df['ateco_sector'].unique()
```

```python
for i in sector_list:
    temp_df=full_df[full_df['ateco_sector'] == i]
    sector_default_rate = len(temp_df[temp_df['default'] == 1])/ len(temp_df)*100
    print('sector:',i,'overall_sector_default_rate:',sector_default_rate,', overall_sector_default:',len(temp_df[temp_df['default'] == 1]),',overall_sector_non_default:',len(temp_df[temp_df['default'] == 0]))
```

<h1># legal Overall default rate

```python
legal_struct_list=full_df['legal_struct'].unique()
```

```python
legal_struct_list
```

```python
for i in legal_struct_list:
    temp_df=full_df[full_df['legal_struct'] == i]
    legal_default_rate = len(temp_df[temp_df['default'] == 1])/ len(temp_df)*100
    print('legal_structure:',i,'overall_legal_default_rate:',legal_default_rate,', overall_legal_default:',len(temp_df[temp_df['default'] == 1]),',overall_legal_non_default:',len(temp_df[temp_df['default'] == 0]))
```


<h1>#yearwise default rate

```python
fs_years_List=full_df['fs_year'].unique()
fs_years_List
```

<h1> #Overall yearwise default rate

```python
for i in fs_years_List:
    temp_df=full_df[full_df['fs_year'] == i]
    year_default_rate = len(temp_df[temp_df['default'] == 1])/ len(temp_df)*100
    print('year:',i,'overall_year_default_rate:',year_default_rate,', overall_year_default:',len(temp_df[temp_df['default'] == 1]),',overall_year_non_default:',len(temp_df[temp_df['default'] == 0]))
```

<h1>#Overall yearly sectorwise default rate

```python
for i in fs_years_List:
    temp_df=full_df[full_df['fs_year'] == i]
    for s in sector_list:
        sector_temp_df=temp_df[temp_df['ateco_sector'] == s]
        sector_default_rate = len(sector_temp_df[sector_temp_df['default'] == 1])/ len(sector_temp_df)*100
        print('year:',i,', sector:',s,', overall_sector_default_rate:',sector_default_rate,', overall_sector_default',len(sector_temp_df[sector_temp_df['default'] == 1]),', overall_sector_non_default:',len(sector_temp_df[sector_temp_df['default'] == 0]))
```

<h1>#Overall yearly legal structurewise default rate

```python
for i in fs_years_List:
    temp_df=full_df[full_df['fs_year'] == i]
    for l in legal_struct_list:
        legal_temp_df=temp_df[temp_df['legal_struct'] == l]
        legal_default_rate = len(legal_temp_df[legal_temp_df['default'] == 1])/ len(legal_temp_df)*100
        print('year:',i,', legal_structure:',l,', overall_legal_structure_default_rate:',legal_default_rate,', overall_legal_structure_default',len(legal_temp_df[legal_temp_df['default'] == 1]),', overall_legal_structure_non_default:',len(legal_temp_df[legal_temp_df['default'] == 0]))
```
